```python
import requests
import pandas as pd

def load(file_path):
    # Load la data des CSV
    return pd.read_csv(file_path)

# tableau des chemins des fichiers + leur noms
files = {
    "../Projet-IA/Data/general_data.csv": "general_data",
    "../Projet-IA/Data/employee_survey_data.csv": "employee_survey_data",
    "../Projet-IA/Data/manager_survey_data.csv": "manager_survey_data",
    "../Projet-IA/Data/in_time.csv": "in_time",
    "../Projet-IA/Data/out_time.csv": "out_time",
}

# Dictionary to store loaded data
loaded_data = {}

# Bouclez sur tout les fichiers, load les data, l'enregistrer dans le dico
for file_path, data_name in files.items():
    loaded_data[data_name] = load(file_path)
```

```python
in_time = loaded_data["in_time"]
out_time = loaded_data["out_time"]

#convertissez les colonnes de dates et heures en objets datetime
for col in out_time.columns[1:]:
    out_time[col] = pd.to_datetime(out_time[col], errors='coerce')
    in_time[col] = pd.to_datetime(in_time[col], errors='coerce')

# Créez un DataFrame pour stocker la différence avec la première colonne inchangée
diff_in_out = pd.DataFrame(out_time.iloc[:, 0])

#liste pour stocker les différences
differences = []

#parcourez les colonnes (à partir de la deuxième colonne)
for col in out_time.columns[1:]:
    # Soustrayez les valeurs correspondantes entre les deux tableaux
    differences.append(out_time[col] - in_time[col])

# Concaténez la liste de différences pour former le DataFrame final
diff_in_out = pd.concat([diff_in_out] + differences, axis=1)

# Affichez le résultat
diff_in_out
```

```python
Mean_in_out = diff_in_out.copy()  # Faire une copie du dataframe pour éviter les erreurs de manipulation

#sélectionnez toutes les colonnes à partir de la deuxième colonne
selected_columns = Mean_in_out.iloc[:, 1:]

# Calculez la moyenne pour chaque ligne
row_means = np.mean(selected_columns, axis=1)

# Ajoutez les moyennes à votre DataFrame
Mean_in_out['row_means'] = row_means

# Affichez le DataFrame mis à jour
Mean_in_out
```

```python
#supprimez les colonnes inutiles pour conservé les heures travaillées par jours en moyenne
Mean_hour_employee = Mean_in_out[['Unnamed: 0', 'row_means']]
Mean_hour_employee
```
